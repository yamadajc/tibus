> Resumen: funcion de tiempo

```python
def daytime (x, pattern):
    """ funcion que devuelve el periodo del dia en el que se produjo el ataque

    Args:
        x (str): string que contiene la fecha del ataque 
        pattern (_type_): patron de busqueda

    Returns:
        np.nan: devuelve un nan en el caso de que no se encuentre el patron
        str : devuelve el estado de la fecha del ataque (Morning, Afternoon, Evening, Night)
    """
    import re
    res = re.findall(pattern, x)
    res="".join(res)
    if res in ["06","07","08","09","10","11", "Morning"]:
        return "Morning"
    elif res in ["12","13","14","15","16","17", "Afternoon"]:
        return "Afternoon"
    elif res in ["18","19","20","21","22","23", "Evening"]:
        return "Evening"
    elif res in ["24","01","02","03","04","05", "Night"]:
        return "Night"
    else:
        return x
```

```python
pattern="^\d{2}|[A-Z][a-z]*"

df["time"]=df["time"].astype(str)
df['time'] = df['time'].apply(lambda x: daytime(x, pattern))
```

Esta información me parece interesante, sin embargo, los datos están muy mal registrados. Hay muchas frases largas que significan lo mismo.
Regex:
- Tiburones definidos por nombres (mayúscula + tiburón):` ([A-Z][a-z]*\sshark) `
- Tiburones definidos por su longitud (en yardas):` \d+']\sshark|\d+'\sshark `
- No confirmado: (no confirmado)|(Inválido)|(Cuestionable)|(sin confirmar)


```python
df["species"]=df["species"].astype(str)
df["species"]
```

```python
list(df["species"])
```

```python
#[A-Z][a-z]*\sshark|\d+'] indica que queremos buscar una palabra que empiece por mayuscula y acabe por shark o un numero seguido de ' y shark, 
#\sshark| indica que queremos buscar una palabra que empiece por un numero seguido de ' y shark , 
#d+'\sshark indica que queremos buscar un numero seguido de ' y shark
pattern= "[A-Z][a-z]*\sshark|\d+']\sshark|\d+'\sshark" 
x = "Tiger shark, 3m"

res = re.findall(pattern, x)
res
```

```python
"".join(res)
```

```python
pattern= "[A-Z][a-z]*\sshark|\d+']\sshark|\d+'\sshark" 
x = "Mako shark, 1.5 m [5'] "

res = re.findall(pattern, x)
res=res.replace("]", "")
res
```

```python
def regeshark (x, pattern):
    """ funcion que devuelve el tipo de tiburon que ha atacado usando regex y el patron de busqueda

    Args:
        x (str): string que contiene el tipo de tiburon
        pattern (str): patron de busqueda

    Returns:
        np.nan: en caso en el que no se encuentre el patron devuelve un nan
        str: devuelve el tipo de tiburon
    """
    res = re.findall(pattern, x)
    res="".join(res)
    res=res.replace("]", "")
    if "shark" in res:
        return res
    else:
        return np.nan
```

```python
pattern= "[A-Z][a-z]*\sshark|\d+']\sshark|\d+'\sshark"
df['species_sorted'] = df['species'].apply(lambda x: regeshark(x, pattern))
```

Tratamiento de la columnas de `age`

En la columna `age` aparecen ciertos valores como strings por lo que necesitamos transformarlos en integer. 

Para los valores nulos podremos estimarlos calculando la media y la std para posteriormente inferir de manera aleatoria valores sin peder informacion. 

```python
df['age'].value_counts(dropna=False).head()
```

```python
df['age'].unique()
```

```python
# fucnion que crea lista de numeros en un string
def age(x):
    """ funcion que crea una lista de numeros en un string

    Args:
        x (str): string a analizar

    Returns:
        list: lista de numeros en un string
    """
    if type(x) != str:
        return np.nan
    else:
        return re.findall("\d+", x)
```

```python
df['age']=df['age'].astype(str)
df['age2']=df['age'].apply(lambda x: age(x))
```

```python
df.head(3)
```

Se observa que tenemos varios años, en esos casos significa que ese registro tiene varias victimas. Una posible solucion seria la de aplicar un `explode` para esos registros.

```python



# generamos una columna para aplicar un explode y mantenemos la columna age2 para comprobar 
df['age3'] = df['age2']
# explode de una columna age3
df  = df.explode("age3")
df
```

```python
# ahora podemos rellenar los valores nulos de la columna age3 haciendo inferencia mediante la normal tras calcular la media y std de la columna age3

# calculamos la media y std de la columna age3
media = df["age3"].astype(float).mean()
std = df["age3"].astype(float).std()
```

```python
# importamos las librerias para generar una normal 
import scipy.stats as stats
```

```python

# generamos un objeto para crear una distribucion normal con la media y std de la columna age3
dist = stats.norm(media, std)
```

```python
# rellenamos los valores nulos de la columna age3 con valores aleatorios generados por la distribucion normal
df["age3"] = df["age3"].apply(lambda x: dist.rvs() if pd.isnull(x) else x)

df
```
